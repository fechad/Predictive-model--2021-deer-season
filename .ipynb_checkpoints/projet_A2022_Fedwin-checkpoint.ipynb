##### Température moyenne annuelle

```julia
Temp = CSV.read("data/meteo.csv", DataFrame)
# On remplace les données manquantes par 0
  n = length(Temp[!, "Mean Temp (°C)"])
    for i = 1:n
        Temp[!, "Mean Temp (°C)"][i] = Temp[!, "Mean Temp (°C)"][i] !== missing ?  Temp[!, "Mean Temp (°C)"][i] : 0.0
    end
# conversion des températures en kelvin, pour ne pas avoir de températures négative
    n = length(Temp[!, "Mean Temp (°C)"])
    for i = 1:n
        Temp[!, "Mean Temp (°C)"][i] = Temp[!, "Mean Temp (°C)"][i] + 273
    end

rename!(Temp, "Date/Time" => "Année") #Changer le nom pour simplifier la lecture 
#sélection des donnée de l'automne, donc de septembre à novembre de chaque années
Temp = Temp[[10 <= Dates.month(Temp.Année[i]) <= 11 for i in 1:length(Temp.Année)], :]

Temp[!,"Année"] = [item == item ? Dates.year(item) : "" for item in Temp[!,"Année"]] #Utilisation de l'année seulement
TempMoy = DataFrame(Année = Int64[], Temp = Float64[])
TempMoy = combine(groupby(Temp, :Année), :"Mean Temp (°C)" => mean∘skipmissing => :"Température")
push!(TempMoy, [2013, 0.0])
Temp2021 = TempMoy.Température[22]
#delete!(TempMoy, [22])
        
```

Voyons si le modèle quadratique détien aussi un pouvoir prédictif significatif sur le nombre de cerfs récoltés.

#### 2.2 Quantité de cerfs récoltés annuellement dans les différentes zones de chasse

On doit d'abord additionner le nombre de cerfs récoltés selon les différents engins de chasse.

```julia
zones = groupby(recolte, :Zone)  # On groupe les données par zone

new_df = DataFrame(Année = Int64[], Zone = Int64[], Cerfs= Int64[])  # On initialise un DataFrame vide

for zone in zones  # Pour chaque zone de chasse,
    
    # pour chaque année,
    # on additionne le nombre de cerfs récoltés selon les différents engins de chasse
    df = combine(groupby(zone, :Année), :Cerfs => sum => :Cerfs)  
    
    df2 = DataFrame(Année = df.Année, 
                    Zone = fill(unique(zone.Zone)[1], size(df, 1)),
                    Cerfs = df.Cerfs) 
    
    append!(new_df, df2)  # On ajoute l'information au DataFrame préinitialisé 
end

sort!(new_df, :Année)
first(new_df, 5)  
new_df
```

#### 2.3 Analyse de l'effets de variables explicatives sur le nombre de cerfs mâles adultes récoltés durant la saison de chasse

```julia
function henryplot(y::Vector{<:Real})

    n = length(y)
    ysorted = sort(y)

    p = ( collect(1:n) .- .5 ) /n

    fd = fit(Normal,y)

    q = quantile.(fd,p)

    plot(x=ysorted, y=q, Geom.point,
    Guide.xlabel("Empirical quantiles"), Guide.ylabel("Estimated quantiles"),
    Theme(discrete_highlight_color=c->nothing),
    Geom.abline(color="red"))


end
```

```julia
"""
    construct_structure(x::Vector{<:Real}, order::Int)

Construction de la matrice de structure du modèle polynomial d'ordre `order` à partir du vecteur `x`.
"""
function construct_structure(x::Vector{<:Real}, order::Int)
    
    X = Array{Float64}(undef, length(x), order+1)
    
    for p in 0:order
       X[:,p+1] = x.^p 
    end
    
    return X
    
end
```

### 3.1 Modèle de régression multiple

#### 3.1.1 Fonction pour trouver le meilleur modèle pour chaque zones

```julia
function get_best_model(dataframes::Any)
    zones = unique(new_df[!, :Zone])
    models = DataFrame(RMSE= Float64[], Formula= Any[], Prediction= Any[])
    for termes in combinations([:Année, :Prix_moyen_Essence])
        modelsCombination = DataFrame(RMSE= Float64[], Zone= Int64[], Prediction= Float64[])
        modelsCombination_df_array = []
        formula = term(:Cerfs) ~ sum(term.(termes))
        for dataframe in dataframes
            modelsCombination_df = DataFrame(RMSE= Float64[], Zone= Int64[], Prediction= Float64[])
            for i = 1:length(zones)
                folds = 4
                kfold = Kfold(nrow(groupby(dataframe, :Zone)[i]), folds)

                rmses = Float64[]
                for folds_id in kfold
                    train_fold = groupby(dataframe, :Zone)[i][folds_id, :]
                    valid_fold = groupby(dataframe, :Zone)[i][Not(folds_id), :]

                    m = glm(formula, train_fold, Normal(), IdentityLink())

                    θ̂ = predict(m, valid_fold)
                    θ̂ = convert.(Int64, [ceil(θ̂[i]) for i in 1:size(θ̂ , 1)])
                    rmse = StatsBase.rmsd(θ̂, valid_fold.Cerfs)

                    push!(rmses, rmse)
                end
                m = glm(formula, groupby(dataframe, :Zone)[i], Normal(), IdentityLink())
                pred = max(ceil(predict(m, DataFrame(Année=[2021], Prix_moyen_Essence = [PrixEssence2021]))[1]), 0)
                push!(modelsCombination_df, [mean(rmses), groupby(dataframe, :Zone)[i].Zone[1] , pred])
            end
            push!(modelsCombination_df_array, modelsCombination_df)
        end
        
        zones = modelsCombination_df_array[1].Zone
        RMSE = 0
        cerfs_recoltes = [0 for i in 1:size(modelsCombination_df_array[1].Zone, 1)]
        for result in modelsCombination_df_array
            RMSE = RMSE + mean(result.RMSE)
            for i in 1:size(modelsCombination_df_array[1].Zone, 1)
                cerfs_recoltes[i] = cerfs_recoltes[i] + result.Prediction[i]
            end
        end
        
        push!(models, [RMSE, formula, DataFrame(Zone = modelsCombination_df_array[1].Zone, Cerfs = cerfs_recoltes)])
    end
    models = sort(models, :RMSE)
    return models
end
```

#### 3.1.2 Détection des meilleurs modèles

```julia
dataframes = [MultipleRegData]
best_models = get_best_model(dataframes)
```

### 3.2 Modèle de régression polynomiale

#### Données pour l'année 2021

```julia
anSquare = 2021^2
JPluie2021 = JourPluieAn.Jours_de_pluie[22] 
JPluieSquare2021 = JPluie2021^2
PrixEssence2021 = essenceMoyAn.Prix_moyen_Essence[25]
PrixEssenceSquare2021 = PrixEssence2021^2
PrixEssenceCubic2021 = PrixEssence2021^3
```

#### 3.2.1 Fonction pour trouver le meilleur modèle pour chaque zones

```julia
function get_best_polynomial_model(dataframes::Any)
    zones = unique(new_df[!, :Zone])
    models = DataFrame(RMSE= Float64[], Formula= Any[], Prediction= Any[])
    for termes in combinations([:Année, :Année_square, :Prix_moyen_Essence, :Prix_moyen_Essence_square, :Prix_moyen_Essence_cubic, :Jours_de_pluie,:Jours_de_pluie_square])
        modelsCombination = DataFrame(RMSE= Float64[], Zone= Int64[], Prediction= Float64[])
        modelsCombination_df_array = []
        formula = term(:Cerfs) ~ sum(term.(termes))
        for dataframe in dataframes
            modelsCombination_df = DataFrame(RMSE= Float64[], Zone= Int64[], Prediction= Float64[])
            for i = 1:length(zones)
                folds = 3
                kfold = Kfold(nrow(groupby(dataframe, :Zone)[i]), folds)

                rmses = Float64[]
                for folds_id in kfold
                    train_fold = groupby(dataframe, :Zone)[i][folds_id, :]
                    valid_fold = groupby(dataframe, :Zone)[i][Not(folds_id), :]

                    m = glm(formula, train_fold, Normal(), IdentityLink())

                    θ̂ = predict(m, valid_fold)
                    θ̂ = convert.(Int64, [ceil(θ̂[i]) for i in 1:size(θ̂ , 1)])
                    rmse = StatsBase.rmsd(θ̂, valid_fold.Cerfs)

                    push!(rmses, rmse)
                end
                m = glm(formula, groupby(dataframe, :Zone)[i], Normal(), IdentityLink())
                pred = max(ceil(predict(m, DataFrame(Année=[2021], Année_square = [anSquare],  Prix_moyen_Essence = [PrixEssence2021], Prix_moyen_Essence_square = [PrixEssenceSquare2021], Prix_moyen_Essence_cubic = [PrixEssenceCubic2021], Jours_de_pluie = [JPluie2021], Jours_de_pluie_square = [JPluieSquare2021]))[1]), 0)
                push!(modelsCombination_df, [mean(rmses), groupby(dataframe, :Zone)[i].Zone[1] , pred])
            end
            push!(modelsCombination_df_array, modelsCombination_df)
        end
        
        zones = modelsCombination_df_array[1].Zone
        RMSE = 0
        cerfs_recoltes = [0 for i in 1:size(modelsCombination_df_array[1].Zone, 1)]
        for result in modelsCombination_df_array
            RMSE = RMSE + mean(result.RMSE)
            for i in 1:size(modelsCombination_df_array[1].Zone, 1)
                cerfs_recoltes[i] = cerfs_recoltes[i] + result.Prediction[i]
            end
        end
        
        push!(models, [RMSE, formula, DataFrame(Zone = modelsCombination_df_array[1].Zone, Cerfs = cerfs_recoltes)])
    end
    models = sort(models, :RMSE)
    return models
end
```

## 5.Analyse de la validité du modèle par rapport à 2020 

On teste l'importance de la régression, soit le pouvoir prédictif des variables explicatives choisies.

#### 5.1 Calcul de SST

```julia
y2020 = an2020.Cerfs
ȳ = mean(y2020)

SST2020 = sum( (y2020 .- ȳ).^2 )
```

#### 5.2 Calcul de SSR et SSE

```julia
# Projection des points sur l'hyperplan de régression
ŷ2020 = prediction2020.Cerfs

# Calcul des erreurs observées
e2020 = y2020 - ŷ2020

# Calcul de la variabilité résiduelle (erreur)
SSE2020 = sum( e2020.^2 )

# Calcul de la variabilité expliquée par la régression

SSR2020 = SST2020 - SSE2020
R²2020 = 1 - SSE2020/SST2020

println("e2020 = $e2020")
println("SSR2020 = $SSR2020")
println("SSE2020 = $SSE2020")
println("R²2020 = $R²2020")
```

#### 5.3 Calcul de la statistique observée permettant de tester l'importance de la régression

```julia
n2020 = length(y2020)

p = size(X,2) - 1  #nombre de variables explicatives

F₀2020 = (n2020-p-1)/p * SSR2020/SSE2020
```

```julia
# Construction de l'objet Distribution
using Distributions
F2020 = FDist(p, n2020-p-1)

# Niveau de risque
α = .05

# Calcul du seuil critique du test
c2020 = quantile(F2020, 1-α)

# Affichage de la région critique
d = layer(x -> pdf(F2017, x), 0, 10)
rc = layer(xintercept =[c], Geom.vline(color="red"))
plot(d,rc, Guide.xlabel("Statistique du test"), Guide.ylabel("Densité"))
```
